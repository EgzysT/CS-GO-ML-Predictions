```python
# IPython magic
%config IPCompleter.greedy=True

# Import other notebooks
import import_ipynb
from DataLoader import *

# Imports
import pandas as pd
import json
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_validate
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics

# Pandas settings
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

```python
# Loading and cleaning bad data lines
dfInitial = loadDataset("dataset_initial", 17)
dfSecond = loadDataset("dataset_second", 6)
dfBoth = dfInitial.append(dfSecond, ignore_index=True)
```

```python
display(dfBoth)
```

```python
df2 = dfBoth.copy()
df2["map"] = df2["map"].astype("category")
df2["mapCodes"] = df2["map"].cat.codes

df2["CT_score"] = pd.to_numeric(df2["CT_score"], downcast="unsigned")
df2["T_score"] = pd.to_numeric(df2["T_score"], downcast="unsigned")
df2["is_overtime"] = df2["is_overtime"].astype("bool")

df2["round_status"] = df2["round_status"].astype("category")
df2["round_status_codes"] = df2["round_status"].cat.codes

df2["CT_total_health"] = pd.to_numeric(df2["CT_total_health"], downcast="unsigned")
df2["T_total_health"] = pd.to_numeric(df2["T_total_health"], downcast="unsigned")
df2["CT_num_alive_players"] = pd.to_numeric(df2["CT_num_alive_players"], downcast="unsigned")
df2["T_num_alive_players"] = pd.to_numeric(df2["T_num_alive_players"], downcast="unsigned")
df2["CT_equip_value"] = pd.to_numeric(df2["CT_equip_value"], downcast="unsigned")
df2["T_equip_value"] = pd.to_numeric(df2["T_equip_value"], downcast="unsigned")
df2["CT_money"] = pd.to_numeric(df2["CT_money"], downcast="unsigned")
df2["T_money"] = pd.to_numeric(df2["T_money"], downcast="unsigned")

df2["CT_has_Defuser"] = df2["CT_has_Defuser"].astype("bool")

df2["bomb_site"] = df2["bomb_site"].astype("category")
df2["bomb_site_codes"] = df2["bomb_site"].astype("category").cat.codes

df2["bomb_dropped"] = df2["bomb_dropped"].astype("bool")

df2["num_active_smokes"] = pd.to_numeric(df2["num_active_smokes"], downcast="unsigned")
df2["num_active_molotovs"] = pd.to_numeric(df2["num_active_molotovs"], downcast="unsigned")

df2["round_winner"] = df2["round_winner"].astype("category")
df2["round_winner_codes"] = df2["round_winner"].cat.codes

df2["fileNumber"] = pd.to_numeric(df2["fileNumber"], downcast="unsigned")
df2["indexInFile"] = pd.to_numeric(df2["indexInFile"], downcast="unsigned")
```

```python
X = df2[['mapCodes', 'CT_score', 'T_score', 'is_overtime', 'round_status_codes', 'round_status_time_left', 'CT_total_health', 'T_total_health', 'CT_num_alive_players', 'T_num_alive_players', 'CT_equip_value', 'T_equip_value', 'CT_money', 'T_money', 'CT_has_Defuser', 'bomb_site_codes', 'bomb_dropped', 'num_active_smokes', 'num_active_molotovs']]
y = df2['round_winner_codes']

#Splitting the train and test data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3) # 70% training / 30% testing data

clf=RandomForestClassifier(n_estimators=100, n_jobs=-1)

clf.fit(X_train,y_train)

y_pred=clf.predict(X_test)

print("Accuracy: ", metrics.accuracy_score(y_test, y_pred))
print("Precision: ", metrics.precision_score(y_test, y_pred))
print("ROC_AUC: ", metrics.roc_auc_score(y_test, y_pred))
```

```python
model = RandomForestClassifier(n_estimators=100, n_jobs=-1)

cv_results = cross_validate(model, X, y, cv=5, scoring=('accuracy', 'precision', 'roc_auc', 'f1'))
```
