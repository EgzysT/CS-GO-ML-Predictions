```python
# IPython magic
%config IPCompleter.greedy=True

# Import other notebooks
import import_ipynb
from DataLoader import *

# Imports
import pandas as pd
import json
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_validate
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.model_selection import StratifiedKFold

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB

from sklearn import metrics

# Pandas settings
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

```python
# Loading and cleaning bad data lines
dfInitial = loadDataset("dataset_initial", 17)
dfSecond = loadDataset("dataset_second", 6)
dfBoth = dfInitial.append(dfSecond, ignore_index=True)
```

```python
df2 = dfBoth.copy()
df2["map"] = df2["map"].astype("category")
df2["mapCodes"] = df2["map"].cat.codes

df2["CT_score"] = pd.to_numeric(df2["CT_score"], downcast="unsigned")
df2["T_score"] = pd.to_numeric(df2["T_score"], downcast="unsigned")
df2["is_overtime"] = df2["is_overtime"].astype("bool")

df2["round_status"] = df2["round_status"].astype("category")
df2["round_status_codes"] = df2["round_status"].cat.codes

df2["CT_total_health"] = pd.to_numeric(df2["CT_total_health"], downcast="unsigned")
df2["T_total_health"] = pd.to_numeric(df2["T_total_health"], downcast="unsigned")
df2["CT_num_alive_players"] = pd.to_numeric(df2["CT_num_alive_players"], downcast="unsigned")
df2["T_num_alive_players"] = pd.to_numeric(df2["T_num_alive_players"], downcast="unsigned")
df2["CT_equip_value"] = pd.to_numeric(df2["CT_equip_value"], downcast="unsigned")
df2["T_equip_value"] = pd.to_numeric(df2["T_equip_value"], downcast="unsigned")
df2["CT_money"] = pd.to_numeric(df2["CT_money"], downcast="unsigned")
df2["T_money"] = pd.to_numeric(df2["T_money"], downcast="unsigned")
df2["CT_num_grenades"] = pd.to_numeric(df2["CT_num_grenades"], downcast="unsigned")
df2["T_num_grenades"] = pd.to_numeric(df2["T_num_grenades"], downcast="unsigned")
df2["CT_num_he"] = pd.to_numeric(df2["CT_num_he"], downcast="unsigned")
df2["T_num_he"] = pd.to_numeric(df2["T_num_he"], downcast="unsigned")
df2["CT_num_flash"] = pd.to_numeric(df2["CT_num_flash"], downcast="unsigned")
df2["T_num_flash"] = pd.to_numeric(df2["T_num_flash"], downcast="unsigned")
df2["CT_num_smokes"] = pd.to_numeric(df2["CT_num_smokes"], downcast="unsigned")
df2["T_num_smokes"] = pd.to_numeric(df2["T_num_smokes"], downcast="unsigned")
df2["CT_num_molly"] = pd.to_numeric(df2["CT_num_molly"], downcast="unsigned")
df2["T_num_molly"] = pd.to_numeric(df2["T_num_molly"], downcast="unsigned")

df2["CT_has_Defuser"] = df2["CT_has_Defuser"].astype("bool")

df2["bomb_site"] = df2["bomb_site"].astype("category")
df2["bomb_site_codes"] = df2["bomb_site"].astype("category").cat.codes

df2["bomb_dropped"] = df2["bomb_dropped"].astype("bool")

df2["num_active_smokes"] = pd.to_numeric(df2["num_active_smokes"], downcast="unsigned")
df2["num_active_molotovs"] = pd.to_numeric(df2["num_active_molotovs"], downcast="unsigned")

df2["round_winner"] = df2["round_winner"].astype("category")
df2["round_winner_codes"] = df2["round_winner"].cat.codes

df2["fileNumber"] = pd.to_numeric(df2["fileNumber"], downcast="unsigned")
df2["indexInFile"] = pd.to_numeric(df2["indexInFile"], downcast="unsigned")
```

```python
# X2 = df2[['mapCodes', 'CT_score', 'T_score', 'is_overtime', 'round_status_codes', 'round_status_time_left', 'CT_total_health', 'T_total_health', 'CT_num_alive_players', 'T_num_alive_players', 'CT_equip_value', 'T_equip_value', 'CT_money', 'T_money', 'CT_num_grenades', 'T_num_grenades', 'CT_num_he', 'T_num_he', 'CT_num_flash', 'T_num_flash', 'CT_num_smokes', 'T_num_smokes', 'CT_num_molly', 'T_num_molly', 'CT_has_Defuser', 'bomb_site_codes', 'bomb_dropped', 'num_active_smokes', 'num_active_molotovs']]
# X = df2.drop(['Debug', 'round_winner_codes', 'round_winner', 'fileNumber', 'indexInFile', 'map', 'round_status', 'bomb_site'], axis=1)
X = df2.drop(['round_winner_codes', 'round_winner', 'fileNumber', 'indexInFile', 'map', 'round_status', 'bomb_site'], axis=1)
# print(list(set(X.columns) - set(X2.columns)) + list(set(X2.columns) - set(X.columns)))
y = df2['round_winner_codes']

#Splitting the train and test data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3) # 70% training / 30% testing data
```

```python
## Pipelines Creation
## 1. Data Preprocessing by using Standard Scaler
## 2. Reduce Dimension using PCA
## 3. Apply  Classifier
```

```python
pipeline_linr = Pipeline([('scalar0',StandardScaler()),
                      #('pca1',PCA(n_components=10)),
                      ('gnb_classifier', GaussianNB())])
```

```python
pipeline_logir = Pipeline([('scalar1',StandardScaler()),
                      #('pca1',PCA(n_components=10)),
                      ('lr_classifier',LogisticRegression(random_state=0, max_iter=1000, n_jobs=-1))])
```

```python
pipeline_dt=Pipeline([('scalar2',StandardScaler()),   # Decision trees and Random Forest do not need scalers, but I'm adding it because why not
                     #('pca2',PCA(n_components=10)),
                     ('dt_classifier',DecisionTreeClassifier())])
```

```python
pipeline_randomforest=Pipeline([('scalar3',StandardScaler()),
                     #('pca3',PCA(n_components=10)),
                     ('rf_classifier',RandomForestClassifier(n_estimators=100, n_jobs=-1))])
```

```python
## Let us make a list of the pipelines
pipelines = [pipeline_linr, pipeline_logir, pipeline_dt, pipeline_randomforest]
```

```python
folds = 5
for pl in pipelines:
    cv = StratifiedKFold(n_splits=folds, shuffle=True)
#     scores = cross_val_score(pl, X_train, y_train, cv = cv)
    scores = cross_validate(pl, X, y, cv=cv, scoring=('accuracy', 'precision', 'roc_auc', 'f1'))
    display(scores)
```

```python
clf=RandomForestClassifier(n_estimators=100, n_jobs=-1)

clf.fit(X_train,y_train)

y_pred=clf.predict(X_test)

print("Accuracy: ", metrics.accuracy_score(y_test, y_pred))
print("Precision: ", metrics.precision_score(y_test, y_pred))
print("ROC_AUC: ", metrics.roc_auc_score(y_test, y_pred))
```

```python
model = RandomForestClassifier(n_estimators=100, n_jobs=-1)

cv_results = cross_validate(model, X, y, cv=5, scoring=('accuracy', 'precision', 'roc_auc', 'f1'))
```
